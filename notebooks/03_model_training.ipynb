# 03 - Model Training

This notebook loads the cleaned training, builds and trains four machine learning models (Logistic Regression, Random Forest, SVM, Gradient Boosting).

This is to bulid/train various suitable models. Model evaluation and performance visualization will be carried out later in the next step.

## Load the Libraries

```python
# Load the required libraries

import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import joblib
```

## Load Cleaned Training Data

Note: After preprocessing, cleaned Train and Test Datasets are saved onto GitHub separately as cleaned_credit_train.csv & cleaned_credit_test.csv

```python
# Load Train Dataset

train_url = "https://raw.githubusercontent.com/KrisAimlGitHub/credit-risk-capstone/main/data/processed/cleaned_credit_train.csv"

train_df = pd.read_csv(train_url)

print('Train DF Shape',train_df.shape)
```

## Identify Features and Target

```python
# Identify X_train and y_train

X_train = train_df.drop('target', axis=1)
y_train = train_df['target']
```

## Model Building

Note: We will build the following 4 models and see their performance.

### Logistic Regression

```python
# Define Model - Logistic Regression
model_lr = LogisticRegression(max_iter=1000)

print("Logistic Regression Model built and trained")
```

### Random Forest

```python
# Define Model - Random FOrest
model_rm = RandomForestClassifier(n_estimators=200, random_state=42)

# Train the model on Training Data
model_rm .fit(X_train, y_train)

print("Random Forest Model built and trained")
```

### Support Vector Classification (SVM/SVC)

```python
# Define Model - SVM
model_sv = SVC(kernel='rbf', probability=True, random_state=42)

# Train the model on Training Data
model_sv .fit(X_train, y_train)

print("SVM Model built and trained")
```

### Gradient Boosting Classifier

```python
# Define Model - Gradient Boosting CLassifier
model_gb = GradientBoostingClassifier(random_state=42)

# Train the model on Training Data
model_gb .fit(X_train, y_train)

print("Gradient Boosting Model built and trained")
```

## Save/Pickle the Models

Let's save pickled models. These will be save in the Colab Env and then manully be uploaded to GitHub

```python
# Save the models to Colab Env

joblib.dump(model_lr, "logistic_regression.pkl")
joblib.dump(model_rm, "random_forest.pkl")
joblib.dump(model_sv, "svm.pkl")
joblib.dump(model_gb, "gradient_boosting.pkl")

print("Models saved successfully.")
```
